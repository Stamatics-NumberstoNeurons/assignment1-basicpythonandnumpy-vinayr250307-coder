```python
# Input — do not modify
def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
```

### Exercise 30 — Softmax Score

**Question:** Compute the softmax score of the sepal length column.

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
sepal_length = iris_2d[:, 0]
def softmax(x):
    e_x = np.exp(x - np.max(x)) # Subtract max for numerical stability
    return e_x / e_x.sum(axis=0)

softmax_scores = softmax(sepal_length)
print(softmax_scores[:5])
```

### Exercise 40 — Numeric to Categorical

**Question:** Bin the petal length (3rd column) of `iris_2d` to form a text array using these rules:
- `< 3` → `'small'`
- `3–5` → `'medium'`
- `>= 5` → `'large'`

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
petal_length = iris_2d[:, 2]

bins = [0, 3, 5, np.inf]
labels = ['small', 'medium', 'large']

# Define conditions based on the bins
conditions = [
    petal_length < bins[1],  # petal_length < 3
    (petal_length >= bins[1]) & (petal_length < bins[2]),  # 3 <= petal_length < 5
    petal_length >= bins[2]   # petal_length >= 5
]

# Use np.select with an explicit string default to avoid the TypeError
petal_length_binned = np.select(conditions, labels, default='unclassified')
print(petal_length_binned[:10])
```

### Exercise 41 — Create New Column

**Question:** Create a new column for **volume** in `iris_2d`, where:

> volume = (π × petal_length × sepal_length²) / 3

*# Make sure you have run Exercise 27 first.*


```python
# Write your code here:
sepal_length = iris_2d[:, 0]
petal_length = iris_2d[:, 2]

volume = (np.pi * petal_length * (sepal_length**2)) / 3
iris_2d_with_volume = np.c_[iris_2d, volume]
print(iris_2d_with_volume[:5])
```

### Exercise 42 — Probabilistic Sampling

### Exercise 43 — Second Largest Value
